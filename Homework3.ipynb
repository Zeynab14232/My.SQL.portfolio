```python
!pip install pandasql
import pandas as pd
import pandasql as psql
Customer=pd.read_excel("https://github.com/ogut77/BA/blob/main/Customer.xlsx?raw=true")
Employee=pd.read_excel("https://github.com/ogut77/BA/blob/main/Employee.xlsx?raw=true")
Employee_1=pd.read_excel("https://github.com/ogut77/BA/blob/main/Employee.xlsx?raw=true")
Product=pd.read_excel("https://github.com/ogut77/BA/blob/main/Product.xlsx?raw=true")
OrdLine=pd.read_excel("https://github.com/ogut77/BA/blob/main/OrdLine.xlsx?raw=true")
OrderTbl=pd.read_excel("https://github.com/ogut77/BA/blob/main/OrderTbl.xlsx?raw=true")
```

```python
#1-List the order number, order date, customer number and customer name (first and last) of orders
# placed in January 2017 sent to Colorado recipients (OrdState).
txt='''
Select OrdNo, OrdDate, Customer.CustNo, CustFirstName, CustLastName
From OrderTbl, Customer
Where OrderTbl.CustNo=Customer.CustNo
and strftime('%Y', date(OrdDate))='2017'
and strftime('%m', date(OrdDate))='01'
and OrdState='CO'

'''
Q1 =  psql.sqldf(txt)
Q1
```

```python
#2-List the order number, order date, and customer number and customer name (first and last) of orders
# placed before January 23, 2017, where customer address (CusState) is in Colorado ('CO')
txt='''
Select OrdNo, OrdDate, Customer.CustNo, CustFirstName, CustLastName
From Customer,OrderTbl
Where Customer.CustNo=OrderTbl.CustNo
and OrdDate<'2017-01-23'
and CustState='CO'
'''
Q2 =  psql.sqldf(txt)
Q2
```

```python
#3-List the order number, order date, customer number and
#customer name (first and last) of orders placed in January 2017 placed by Colorado customers
#(CustState) but sent to Washington recipients (OrdState).
txt='''

Select OrdNo, OrdDate, Customer.CustNo, CustFirstName, CustLastName, OrdState
From Customer, OrderTbl
Where Customer.CustNo=OrderTbl.CustNo
and strftime('%Y', date(OrdDate))='2017'
and strftime('%m', date(OrdDate))='01'
and CustState='CO'
and OrdState='WA'

'''
Q3 =  psql.sqldf(txt)
Q3
```

```python
#3-List the order number, order date, customer number and
#customer name (first and last) of orders placed in January 2017 placed by Colorado customers
#(CustState) but sent to Washington recipients (OrdState).
txt='''

Select OrdNo, OrdDate, Customer.CustNo, CustFirstName, CustLastName, OrdState
From Customer, OrderTbl
Where Customer.CustNo=OrderTbl.CustNo
and strftime('%Y', date(OrdDate))='2017'
and strftime('%m', date(OrdDate))='01'
and CustState='CO'
and OrdState='WA'

'''
Q3 =  psql.sqldf(txt)
Q3
```

```python
#4-List the customer number, name (first and last), and balance of Washington
#customers who have placed  one or more orders in February 2017.
txt='''
Select distinct(Customer.CustNo), CustFirstName, CustLastName, CustBal
From Customer, OrderTbl
Where Customer.CustNo=OrderTbl.CustNo
and CustState='WA'
and strftime('%Y', date(OrdDate))='2017'
and strftime('%m', date(OrdDate))='02'
Group by Customer.CustNo
Having count(OrdNo)>=1

'''
Q4 =  psql.sqldf(txt)
Q4
```

```python
#5-List the customer number, name (first and last), and balance of Washington
#customers who have placed  more than 1 order in February 2017.
txt='''
Select distinct(Customer.CustNo), CustFirstName, CustLastName, CustBal, count(OrdNo)
From Customer, OrderTbl
Where Customer.CustNo=OrderTbl.CustNo
and CustState='WA'
and strftime('%Y', date(OrdDate))='2017'
and strftime('%m', date(OrdDate))='02'
Group by Customer.CustNo
Having count(OrdNo)>1

'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#6-List product related information if product price is lower than average price
txt='''
Select *
From Product
Where ProdPrice< (Select avg(ProdPrice) From Product)

'''
Q6 =  psql.sqldf(txt)
Q6
```

```python
#7-if product price is lower than average price, find how many times are they ordered?
#List product price as well.
txt='''
Select Product. ProdNo, count(OrdNo), ProdPrice
From Product, OrdLine
Where Product.ProdNo=OrdLine.ProdNo
and ProdPrice< (Select avg(ProdPrice) From Product)
Group by Product.ProdNo

'''
Q7 =  psql.sqldf(txt)
Q7
```

```python
#8-List the order number and total amount for orders placed each day.

txt='''
Select OrdDate, Count(OrdLine.OrdNo) as 'Number of Orders', Sum(Qty*ProdPrice) as 'Total Amount'
From OrderTbl, OrdLine, Product
Where OrderTbl.OrdNo=OrdLine.OrdNo
and OrdLine.ProdNo=Product.ProdNo
Group by OrdDate



'''
Q8 =  psql.sqldf(txt)
Q8
```

```python
#9-What is the total value of product sold by each employee?
#Included internet orders as well
txt='''
SELECT OrderTbl.EmpNo, SUM(Qty * ProdPrice) AS TotalSales
FROM OrderTbl
JOIN OrdLine ON OrderTbl.OrdNo = OrdLine.OrdNo
JOIN Product ON OrdLine.ProdNo = Product.ProdNo
LEFT JOIN Employee ON OrderTbl.EmpNo = Employee.EmpNo
GROUP BY OrderTbl.EmpNo


'''
Q9=  psql.sqldf(txt)
Q9
```

```python
#10-List the total commission earned by each employee in each month of 2017.
txt='''
SELECT Employee.EmpNo, OrdDate, Sum(Qty*ProdPrice*EmpCommRate) AS 'Comission Earned'
From OrderTbl, OrdLine,Product,Employee
Where OrderTbl.OrdNo=OrdLine.OrdNo
and OrdLine.ProdNo=Product.ProdNo
and OrderTbl.EmpNo=Employee.EmpNo
and strftime('%Y', date(OrdDate))='2017'
Group by Employee.EmpNo

'''
Q10 =  psql.sqldf(txt)
Q10
```

```python
#11-List the customer first and last name and order number and employee first name and last name
#who handled that order where order is given in February 2017.
#.If order is internet order, include them as well
#Note:If order is internet order, employee first name and last name will be null
txt='''
Select CustFirstName, CustLastName, OrdNo, EmpFirstName, EmpLastName
From Customer
Join OrderTbl on Customer.CustNo=OrderTbl.CustNo
Left Join Employee on Employee.EmpNo=OrderTbl.EmpNo
Where strftime('%Y', date(OrdDate))='2017'
and strftime('%m', date(OrdDate))='02'


'''
Q11=  psql.sqldf(txt)
Q11
```

```python
#12-List the total commission earned by each employee. Includeemployee first and last name,
#supervisor last name and first name if any.If employee does not have supervisor, supervisor last name and first name will be null.
txt='''
Select Employee.EmpNo, Employee.EmpFirstName, Employee.EmpLastName, sum(ProdPrice*Qty*Employee.EmpCommRate) as 'Comission Earned', Employee_1.EmpFirstName AS SupFirstName,Employee_1.EmpLastName AS SupLastName
From Employee Left Join Employee_1
On Employee.SupEmpNo=Employee_1.EmpNo
Join OrderTbl on OrderTbl.EmpNo=Employee.EmpNo
Join OrdLine on OrderTbl.OrdNo=OrdLine.OrdNo
Join Product on OrdLine.ProdNo=Product.ProdNo

'''
Q12 =  psql.sqldf(txt)
Q12
```

```python
#13- List product information which is not ordered?
#Note: if here is not such product, nothing other than column name will be displayed.
txt='''
Select Distinct ProdNo as 'Product'
From Product
Except
Select Distinct ProdNo
From OrdLine

'''
Q13=  psql.sqldf(txt)
Q13
```

```python
#14-How many times each product are ordered? What is the total value of orders taken by each product?
#Whatis the total quantity ordered for each product?
txt='''
Select distinct Product.ProdNo, count(OrdNo) as 'Number of Orders', sum(Qty*ProdPrice)
From Product, OrdLine
Where Product.ProdNo=OrdLine.ProdNo
Group by Product.ProdNo


'''
Q14 =  psql.sqldf(txt)
Q14
```

```python
#15. How many product are there without ProdNextShipDate.
#What is their average price?
txt='''

Select count(*) as 'Number of Product without ProdNextShipDate', avg(ProdPrice) as 'Avergae Price'
From Product
Where ProdNextShipDate is null
'''
Q15 =  psql.sqldf(txt)
Q15
```
